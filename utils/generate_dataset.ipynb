```python
import pandas as pd
import requests
import os

def download_sdss_image(ra, dec, output_dir, scale=0.1, width=256, height=256):
    url = f"http://skyserver.sdss.org/dr16/SkyServerWS/ImgCutout/getjpeg?ra={ra}&dec={dec}&scale={scale}&width={width}&height={height}"
    response = requests.get(url)
    if response.status_code == 200:
        # Garante que o diretório de saída existe
        os.makedirs(output_dir, exist_ok=True)
        filename = os.path.join(output_dir, f"sdss_image_{ra}_{dec}.jpg")
        with open(filename, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded: {filename}")
    else:
        print(f"Failed to download image at RA: {ra}, DEC: {dec}")

# Defina o diretório onde as imagens serão salvas
output_dir = r"..\data\skyserver\dr18\120_0.2"

# Ler o CSV corretamente
csv_path = r"..\data\skyserver\dr18\clean_fotometry.csv"
df = pd.read_csv(csv_path)  # Certifique-se de ler o arquivo CSV para um DataFrame

# Iterar sobre as linhas do DataFrame e baixar as imagens
for index, row in df.iterrows():
    ra = row['ra']
    dec = row['dec']
    download_sdss_image(ra=ra, dec=dec, output_dir=output_dir, scale=0.2, width=120, height=120)
```
